```python
#import all required packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#read in data from csv
deaths = pd.read_csv(r"/Users/trinitybibler/Documents/drug_deaths.csv")
```

```python
#create numpy arrays and reshape into 2d arrays as 1d arrays do not work with algorithm
x = deaths['State'].values.reshape(-1,1)
y = deaths['deaths_20'].values.reshape(-1,1)

```

```python
#must fit the data before predictions can be made
reg = LinearRegression()
reg.fit(x, y)
```

```python
#add predictions
death_pred = reg.predict(x)
```

```python
#show scatter plot of actual data versus predicted data
plt.scatter(x, y, color = 'gray', label = 'actual')
plt.plot(x, death_pred, color = 'red', linewidth = 2, label = 'predicted')
plt.legend()
plt.xlabel('State')
plt.ylabel('Deaths')
plt.title('Actual versus Predicted by state')
plt.show()
```
